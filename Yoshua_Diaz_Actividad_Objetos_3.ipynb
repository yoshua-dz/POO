```python
from IPython.display import Image, display

class EmpleadoUniversitario:
  def __init__(self, nombre, departamento, salario, foto=None):
    self._nombre = nombre
    self._departamento = departamento
    self._salario = salario
    self._foto = foto

  def mostrar_informacion(self):
    print("\nNombre: ", self.nombre)
    print("Departamento: ", self.departamento)
    print("Salario: ", self.salario)

  @property #El atributo solo se le será permitido leerse // getters
  def nombre(self):
    return self._nombre

  @property
  def departamento(self):
    return self._departamento

  @property
  def salario(self):
    return self._salario

  @property
  def foto(self):
    return self._foto

  # @atributo.setter nos ayuda al redefinir los valores de los atributos // setters
  @nombre.setter
  def nombre(self, nuevo_nombre):
    self._nombre = nuevo_nombre

  @departamento.setter
  def departamento(self, nuevo_departamento):
    self._departamento = nuevo_departamento

  @salario.setter
  def salario(self, nuevo_salario):
    if nuevo_salario < 0:
      print("Error: El salario no puede ser negativo.")
    else:
      self._salario = nuevo_salario

  @foto.setter
  def foto(self, nueva_foto):
    self._foto = nueva_foto

  def mostrar_foto(self): #impresión de la imagen
    if self.foto:
      display(Image(self.foto))
    else:
      print("No hay foto disponible.")


empleado1 = EmpleadoUniversitario("Henry Cavill", "Ventas", 5000, "https://m.media-amazon.com/images/M/MV5BNWFmNmI2NmYtNDUyYi00MDZhLTk0ZmEtODY0MmM5NGQxMWEwXkEyXkFqcGc@._V1_FMjpg_UX1000_.jpg")

empleado1.mostrar_informacion()
empleado1.mostrar_foto()
empleado1.salario = -6000
empleado1.departamento = "Marketing"
empleado1.nombre = "Tom Holland"
empleado1.foto = "https://upload.wikimedia.org/wikipedia/commons/3/3c/Tom_Holland_by_Gage_Skidmore.jpg"
empleado1.mostrar_informacion()
empleado1.mostrar_foto()
```
